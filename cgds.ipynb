### **2. Logistique : Optimisation des cha√Ænes d'approvisionnement et gestion des stocks**

#### **Optimisation des cha√Ænes d'approvisionnement**
- **Probl√®me** : Une entreprise de distribution veut minimiser les co√ªts de transport tout en respectant les d√©lais de livraison.
- **Solution** :
  - Utilisation de l'optimisation lin√©aire pour d√©terminer les routes les plus efficaces.
  - Donn√©es utilis√©es : Co√ªts de transport, distances, d√©lais de livraison, etc.
  - Algorithme : Programmation lin√©aire (avec la biblioth√®que PuLP en Python).
  - **Exemple de code Python** :
    ```python
    from pulp import LpMinimize, LpProblem, LpVariable

    # D√©finition du probl√®me
    prob = LpProblem("Optimisation_Transport", LpMinimize)

    # Variables de d√©cision
    x1 = LpVariable("Route_1", lowBound=0, cat='Integer')  # Quantit√© √† transporter par la route 1
    x2 = LpVariable("Route_2", lowBound=0, cat='Integer')  # Quantit√© √† transporter par la route 2

    # Fonction objectif (minimiser les co√ªts)
    prob += 50 * x1 + 70 * x2, "Co√ªt_total"

    # Contraintes
    prob += x1 + x2 == 100, "Demande_totale"
    prob += x1 <= 60, "Capacit√©_Route_1"
    prob += x2 <= 50, "Capacit√©_Route_2"

    # R√©solution
    prob.solve()
    print(f"Quantit√© √† transporter par la route 1 : {x1.varValue}")
    print(f"Quantit√© √† transporter par la route 2 : {x2.varValue}")
    ```

#### **Gestion des stocks**
